# Histograms
Histograms reveal, through numbers, interpretable trends in your data. They donâ€™t provide a yes or no answer, but are often used as a starting point for discussion and informing an answer to your data.

The purpose of a histogram is to summarize data that you can use to inform a decision or explain a distribution.

### Choose a correct number of bins
We need that all the bars of the histogram have the same width. To calculate this we will use the ruler known as Sturges rule.

#### Sturges rule
To obtain the Sturges rule, it is only necessary to assume that the number of samples in each interval of an ideal histogram is given by the binomial coefficient. Thus, the total number of samples can be obtained by adding the number of each of the intervals.

$${N} = {\sum_{i}^{k-1}\binom{k-1}{i} = 2^{k-1}}$$

Therefore, we can obtain a formula that depends only on the number of values that exist in the data set.

$${K_{sturges}} = \log_2{|n|} + 1$$

### Calculating Mean

The mean, often referred to as the average, is a way to measure the center of a dataset.

The average of a set is calculated using a two-step process:

1. Add all of the observations in your dataset.
2. Divide the total sum from step one by the number of points in your dataset.

$$\bar{x} = \frac{x_1 + x_2 â€¦ + x_{n}}{n} $$

```python
#calculate the average of sales, expenses and utilities
average_sales = int(np.average(sales))
average_expenses = int(np.average(expenses))
average_utilities = int(np.average(utilities))
```

```python
#Graph the average of sales in 2021
print(f"The average of sales in 2021 of the company was: ${average_sales}")
plt.hist(sales, range = (min_valueS, max_valueS), bins=bins, edgecolor='black')
plt.title("Histogram of sales of the company in 2021")
plt.xlabel("Value")
plt.ylabel("Freq")
plt.axvline(average_sales, color='r', linestyle='solid', linewidth=2, label="Sales average")
plt.legend()
plt.show()
```

```python
#Graph the average of expenses in 2021
print(f"The average of expenses in 2021 of the company was: ${average_expenses}")
plt.hist(expenses, range = (min_valueE, max_valueE), bins=bins, edgecolor='black')
plt.title("Histogram of expenses of the company in 2021")
plt.xlabel("Value")
plt.ylabel("Freq")
plt.axvline(average_expenses, color='r', linestyle='solid', linewidth=2, label="Expenses average")
plt.legend()
plt.show()
```

## Median

The formal definition for the median of a dataset is:

_The value that, assuming the dataset is ordered from smallest to largest, falls in the middle. If there are an even number of values in a dataset, you either report both of the middle two values or their average._

There are always two steps to finding the median of a dataset:

1. Order the values in the dataset from smallest to largest
2. Identify the number(s) that fall(s) in the middle

```python
#calculate the median of the sales, expenses and utilities
median_sales = int(np.median(sales))
median_expenses = int(np.median(expenses))
median_utilities = int(np.median(utilities))
```

```python
#Graph the median of sales in 2021
print(f"The average of sales per month in 2021 of the company was: ${average_sales}")
print(f"The median of sales per month in 2021 of the company was: ${median_sales}")
plt.hist(sales, range = (min_valueS, max_valueS), bins=bins, edgecolor='black')
plt.title("Histogram of sales of the company in 2021")
plt.xlabel("Value - $")
plt.ylabel("Freq")
plt.axvline(median_sales, color='orange', linestyle='solid', linewidth=2, label="Sales median")
plt.axvline(average_sales, color='r', linestyle='solid', linewidth=2, label="Sales average")
plt.legend()
plt.show()
```

```python
#Graph the median of expenses in 2021
print(f"The average of expenses in 2021 of the company was: ${average_expenses}")
print(f"The median of expenses in 2021 of the company was: ${median_expenses}")
plt.hist(expenses, range = (min_valueE, max_valueE), bins=bins, edgecolor='black')
plt.title("Histogram of expenses of the company in 2021")
plt.xlabel("Value - $")
plt.ylabel("Freq")
plt.axvline(median_expenses, color='orange', linestyle='solid', linewidth=2, label="expenses median")
plt.axvline(average_expenses, color='r', linestyle='solid', linewidth=2, label="expenses average")
plt.legend()
plt.show()
```

## Mode

The formal definition for the mode of a dataset is:

_The most frequently occurring observation in the dataset. A dataset can have multiple modes if there is more than one value with the same maximum frequency._

There is not directly method to find the mode using Numpy, for that reaseon we will use SciPy fot this task.

```python
#calculate the mode of sales, expenses and utilities
mode_sales = stats.mode(sales)[0]
mode_expenses = stats.mode(expenses)[0]
mode_utilities = mode(utilities)
```

# Variance
Variance is a descriptive statistic that describes how spread out the points in a data set are.

The full equation for the variance is as follows:

$$\sigma^2 = \frac{\sum_{i=1}^{N}{(X_i -\mu)^2}}{N}$$


```python
#calculate the variance of sales, expenses and utilities
variance_sales = int(np.var(sales))
variance_expenses = int(np.var(expenses))
variance_utilities = int(np.var(utilities))
```

```python
#Graph the histogram
plt.hist(sales, range = (min_valueU, max_valueS), bins=bins, edgecolor='black', alpha=0.80, label="Sales")
plt.hist(expenses, range = (min_valueU, max_valueS), bins=bins, edgecolor='black', alpha=0.6, label="Expenses")
plt.hist(utilities, range = (min_valueU, max_valueS), bins=bins, color="green", edgecolor='black', alpha=0.4, label="Utilities")
plt.title("Histogram of sales, expenses and utilities of the company in 2021")
plt.xlabel("Value ~ $")
plt.ylabel("Freq")
plt.legend()
plt.show()

print(f"The average of sales is: ${average_sales}, and the variance is: {variance_sales}")
print(f"The average of expenses is: ${average_expenses}, and the variance is: {variance_expenses}")
print(f"The average of utilities is: ${average_utilities}, and the variance is: {variance_utilities}")
print("We observe a very large variance in each measures of central tendency.")
print("The cause of this is because the data is too away from each other and all the data is in order of (10^6).")
```

## Standard Deviation

Standard deviation is computed by taking the square root of the variance. `sigma` is the symbol commonly used for standard deviation. Conveniently, `sigma` squared is the symbol commonly used for variance:

$$\sigma = \sqrt{\sigma^2} = \sqrt{\frac{\sum_{i=1}^{N}{(X_i -\mu)^2}}{N}}$$

```python
#we calculate the standard deviatio of sales, expenses and utilities of the company in 2021
std_sales = np.std(sales)
std_expenses = np.std(expenses)
std_utilities = np.std(utilities)
print(f"The standard deviation of all the data of the sales of the company is: {std_sales}")
print(f"The standard deviation of all the data of the expenses of the company is: {std_expenses}")
print(f"The standard deviation of all the data of the utilities of the company is: {std_utilities}")
print("\nWe observe a deviation smaller than the variance, but it's still very large, because the data is in order of the (*10^7)")
```

## Boxplots in Matplotlib

The `matplotlib.pyplot` module has a function named `boxplot()`. `boxplot()` takes a dataset as a parameter. This dataset could be something like a list of numbers, or a Pandas DataFrame.

Some of the major takeaways from boxplots are:

- The box of a boxplot visualizes the median, first quartile, and third quartile of a dataset.
- The length of the box in a boxplot visualizes the interquartile range.
- The whiskers extend from the box 1.5 times the size of the interquartile range.
- Outliers are points that fall outside of the whiskers. Theyâ€™re represented by dots.
- Boxplots are especially useful for comparing the spread of multiple datasets.

```python
plt.boxplot([sales, expenses, utilities])
plt.title("Boxplots of sales, expenses and utilities of the company in 2021\n")
plt.show()
```
